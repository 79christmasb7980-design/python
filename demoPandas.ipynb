```python
#딕셔너리로 저장
data ={"foreign" : [1,2,3,4,5],
       "sratio" : [10,20,30,40,50],
       "org" : [100,200,300,400,500],
       "sprice" : [1,2,3,4,5],
       "private" : [90,180,270,360,450]
       }
#데이터프레임으로 로딩
frame1 = DataFrame(data)
frame1
```

```python
#컬럼의 순서 변경
frame2 = DataFrame(data, columns=["sprice","private","foreign","sratio","org"],
                   index=['2025-11-14', '2025-11-15', '2025-11-16', '2025-11-17', '2025-11-18'])
#컬럼을 슬라이싱
frame2["org"]
fram3 = frame2[["org","sratio"]]
fram3
```

```python
#loc 인덱서 : 라벨기반 인덱싱
frame4 = DataFrame(frame2.loc['2025-11-15':'2025-11-17', ['sprice','org']])
frame4
```
